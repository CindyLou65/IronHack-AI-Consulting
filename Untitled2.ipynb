```python
# Your answer here
#Pseudocode
#Ask the user to enter 1 or 2
#2. Read input into choice
#3. WHILE choice is not 1 and not 2:
       Ask the user again
       Read input again
#4. IF choice is 1:
       Convert Fahrenheit to Celsius
#5. ELSE (choice is 2):
       Convert Celsius to Fahrenheit

#1. Display "Enter 1 to convert from Fahrenheit to Celsius or 2 to convert from Celsius to Fahrenheit:"
#2. Read user input and assign it to the variable 'choice'
#3. While the choice is not 1 and is not 2 then ask for user input again.
#4. Read user input again
#5. If 'choice' is equal to 1, go to step 6. If choice is equal to 2, go to step 9.
```

```python
#### My code here:
# Prompt the user to choose a conversion type

choice = int(input("Press 1 to convert from Fahrenheit to Celsius or Press 2 to convert from Celsius to Fahrenheit: "))

# Based on the choice, ask for the temperature and perform the conversion
while choice != 1 and choice != 2:
    print("Must be number 1 or 2. Try again!")
    choice = int(input("Press 1 to convert from Fahrenheit to Celsius or Press 2 to convert from Celsius to Fahrenheit: "))
    
if choice == 1:
    fahrenheit = float(input("Enter the temperature in Fahrenheit: "))
    celsius = (fahrenheit - 32) * 5 / 9
    print(f"The converted temperature is {celsius:.2f} degrees Celsius.")
    
elif choice == 2:
    celsius = float(input("Enter the temperature in Celsius: "))
    fahrenheit = (celsius * 9 / 5) + 32
    print(f"The converted temperature is {fahrenheit:.2f} degrees Fahrenheit.")
    
else:
    print("Invalid choice. Please enter either 1 or 2.")
```
