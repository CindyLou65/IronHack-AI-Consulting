```python
# Introduction to Functions Exercise
#Create a function named "generation_calculator" to help determine the generation you belong to according to your birth year. Follow the below information:

#Birth Year	Generation
#< 1949	Silent Generation
#1949-1968	Baby Boomer
#1969-1980	Generation X
#1981-1993	Millennial
#>= 1994	Generation Z

#Define the function generation calculator. Write conditionals for each generation. For example, if you were born before 1949 you would be a "silent generation".
#Call the function to test it with different year values.
#Print the results.
```

```python
birth_year = int(input("Please input your birth year"))
def generation_calculator(birth_year):
    
    if birth_year < 1949:
        print("You are from the Silent Generation")

    elif 1949 <= birth_year <= 1968:
        print("You are a Baby Boomer")

    elif 1969 <= birth_year <= 1980:
        print("You are from Generation X")

    elif 1981 <= birth_year <= 1993:
        print("You are a Millenial")

    else:
        print("You are Generation Z")

generation_calculator(birth_year)
```
