```python
#Lab M2.03_Python Functions_Cindy_Lund

#Exercise 1: Create a Greeting Function (Easy - From Scratch)
#Scenario: You want to create a function that greets people by name and can optionally include their favorite programming language.

#What to do:
#Create a function called greet_person that takes two parameters:
#name (a string): The person's name
#language (a string, optional): Their favorite programming language (default to "Python" if not provided)
#Inside the function, create a greeting message that says:

#"Hello, [name]! Welcome to Python programming!"
#If a language is provided (and it's not "Python"), also add: "I see you're interested in [language] too!"
#Return the greeting message as a string

#Test your function with:

#greet_person("Alice") (should use default language)
#greet_person("Bob", "JavaScript") (should mention JavaScript)
#greet_person("Charlie", "Python") (should not mention the language since it's the default)
#Hints:

#Use def to define your function
#Use default parameter values: language="Python"
#Use conditional statements (if) to check if the language is provided and different from "Python"
#Use string formatting or concatenation to build your message
#Use return to send back the message
```

```python
name = input("Please input your name ")
language = input("Please input your favorite programming language " )

def greet_person (name, language):
    print("Hello ", name,"! Welcome to Python programming!")

    if not language:
        return

    elif language == "Python":
        return

    elif language == "python":
        return

    elif language != "Python":
        print("I see you're interested in", language, "too!")

greet_person(name, language)     
```

```python
percentage = float(input("Enter your percentage: "))
def calculate_grade(percentage):
    if percentage >= 90:
        grade = 'A'
    elif percentage >= 80:
        grade = 'B'
    elif percentage >= 70:
        grade = 'C'
    elif percentage >= 60:
        grade = 'D'
    else:
        grade = 'F'

    print('Your grade is', grade)

calculate_grade(percentage)
```

```python
#Bonus: Bonus Challenge: Add input validation to check if the percentage is between 0 and 100. 
#If it's not, return an error message like "Invalid percentage. Please enter a value between 0 and 100."

percentage = float(input("Enter your percentage: "))
    
def calculate_grade(percentage):
    if percentage < 0 or percentage > 100:
        return "Invalid percentage. Please enter a value between 0 and 100."
    
    elif percentage >= 90:
        grade = 'A'
    elif percentage >= 80:
        grade = 'B'
    elif percentage >= 70:
        grade = 'C'
    elif percentage >= 60:
        grade = 'D'
    else:
        grade = 'F'

    print('Your grade is', grade)

calculate_grade(percentage)
```
