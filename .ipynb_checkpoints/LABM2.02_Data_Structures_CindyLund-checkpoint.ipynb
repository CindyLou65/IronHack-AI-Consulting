```python
# EXERCISES
# Working with Lists
# You're creating a guest list for your event. Lists are perfect for this because they maintain order and allow duplicates (in case someone RSVPs twice).

#What to do:
#Create a list called guest_list with at least 5 guest names (you can use: "Alice", "Bob", "Charlie", "Diana", "Eve")
#Add two more guests to the end of the list using .append()
#Print the total number of guests using len()
#Print the first guest and the last guest using indexing
#Sort the list alphabetically using .sort()
# Print the sorted list
```

```python
guest_list = ["Alice", "Bob", "Charlie", "Diana", "Eve"]
guest_list.append("Paul")
guest_list.append("Mike")
print(len(guest_list))

print(guest_list[0])
print(guest_list[-1])

guest_list.sort()
print(guest_list)

```

```python
# Exercise 2: Working with Dictionaries
# Scenario: You need to store contact information for each guest. Dictionaries are perfect because you can quickly look up a guest's email by their name.

#What to do:

# Create a dictionary called contact_info with at least 3 key-value pairs where:
# Keys are guest names (strings)
# Values are their email addresses (strings)
# Example: {"Alice": "alice@email.com", "Bob": "bob@email.com", "Charlie": "charlie@email.com"}
# Add 2 more guests and their emails to the dictionary
# Print the email address of one specific guest by using their name as the key
# Print all the keys (guest names) using .keys()
# Print all the values (email addresses) using .values()
# Check if a specific guest exists in the dictionary using the in keyword
```

```python
contact_info = {"Alice": "alice@email.com", "Bob": "bob@email.com", "Charlie": "charlie@email.com"}
contact_info["Mike"] = "Mike@email.com"
contact_info["Julie"] = "Julie@email.com"
print(contact_info["Mike"])
print(contact_info.keys())
print(contact_info.values())
if "Bob" in contact_info:
    print(contact_info['Bob'])

else:
    print("Bob is not on the list")


```

```python
#Exercise 3: Working with Sets
#Scenario: You want to collect unique interests from your guests to plan activities. Sets are perfect because they automatically remove duplicates and make it easy to find common interests.

#What to do:

#Create a set called interests with at least 5 different interests (e.g., "music", "reading", "sports", "cooking", "travel")
#Add 2 more interests using .add()
#Try adding a duplicate interest (one that's already in the set) and observe what happens
#Print the set and notice that it doesn't maintain order
#Create a second set with some overlapping interests
#Find the common interests between the two sets using .intersection() or the & operator
#Print the result
```

```python
interests = {"music", "reading", "sports", "cooking", "travel"}
interests.add("swimming")
interests.add("sightseeing")
interests.add("music")
print(interests)
interests_new = {"reading", "sports", "dancing", "theater"}
print(interests.intersection(interests_new))

```
